### Drop Columns: Eliminar `columnas` de un dataset

Usamos la funcion `.drop`: recibe como parametros de entrada el label de la columna, el eje (axis = 0 --> filas) (axis = 1--> columnas)

```python
df_books.drop('Genre',axis=1).head(2)
```

```python
df_books.head(2)
```

Es muy importante ver que `.drop( )` **NO** modifica el dataframe original, sino unicamente en la salida o en una nueva variable


Si reescribimos el dataframe tambien es posible borrar la columna

```python
df_books= df_books.drop('Year',axis=1).head(2)
```

```python
df_books
```

Si queremos que se borre del dataframe sin reescribir el df

```python
df_books.drop('Price',axis=1, inplace=True)
```

```python
df_books
```

```python
del df_books['Reviews']
```

```python
df_books.head(2)
```

### Drop rows: Eliminar `Filas`

Aplica igual que en columnas, el parametro `inplace`, sobreescribir el dataframe

```python
df_books = pd.read_csv('/home/diemtca/personalProjects/DS_proyects/Learn_Pandas/bestsellers-with-categories.csv',sep=',',header=0)
```

```python
df_books.head(2)
```

```python
df_books.drop(0,axis=0).head(2)
```

Podemos eliminar varias filas a manera de lista

```python
df_books.drop([0,1,2],axis=0).head(2)
```

Ahora lo haremos a manera de lista

```python
df_books.drop(range(0,10),axis=0).head(2)
```

Ya sabemos borrar filas y columnas, pero ¿como agregamos filas o Columnas?

### Add `Columns`: Agregar `Columnas`

```python
df_books.head()
```

Podemos hacerlo de manera genérica

```python
df_books['Nueva Columna'] = np.nan
df_books.head(2)
```

Para agregar una `columna` de esta manera es importante asegurarse que el numero de datos de la nueva `columna` debe ser el mismo que el numero de filas del dataframe

```python
#pregunto cuantas filas tiene el dataframe
df_books.shape[0]
```

```python
#creamos datos con el tamaño de las filas del df
data = np.arange(0,df_books.shape[0])
```

```python
#agregamos una nueva columna con los datos creados
df_books['Rango'] = data
df_books
```

### Add `Rows` : Agregar `Filas`

```python
df_books.append(df_books)
```

Se puede observar que se duplicaron los registros, ahora el dataframe tiene 1100 filas y nueve columnas, lo confirmamos con el metodo ``.shape[:]``

```python

df_books.append(df_books).shape[:]
```

Otra forma de hacerlo con el metodo `pd.concat([df_1, df_2])`

```python
pd.concat([df_books,df_books])
```

```python
pd.concat([df_books,df_books]).shape[:]
```
